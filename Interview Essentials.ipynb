# 4. To print Fibonacci series

```python
def fib(n):
    a=0
    b=1
    
    if(n==1):
        print(a)
        
    else:
        print(a)
        print(b)
        
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            print(c)
fib(int(input()))
```

```python
# if n is negative
def fib(n):
    a=0
    b=1
    
    if(n==1):
        print(a)
        
    else:
        print(a)
        print(b)
        
        for i in range(2,n):
            c=a+b
            a=b
            b=c
            print(c)
            
n=int(input('Enter the range of series'))

if(n<0):
    print('Please enter a valid number greater than zero')
else:
    fib(n)
```

# 8. To print #'s in increasing order

```python
for i in range(4):
    for j in range(i+1):
        print('#',end=" ")
        
    print()
        
```

# 8. To print #'s in decreasing order

```python
for i in range(4):
    for j in range(4-i):
        print('#',end=" ")
        
    print()
```
