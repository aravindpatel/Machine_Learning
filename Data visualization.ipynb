```python
t.drop('Cabin',axis=1,inplace=True)
```

```python
t.dropna(inplace=True)
```

```python
sns.heatmap(t.isnull(),yticklabels=False,cmap='viridis')
```

```python
t.isnull().sum()
```

```python
t.head()
```

```python
sex=pd.get_dummies(t['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
embark=pd.get_dummies(t['Embarked'])
```

```python
embark.head(5)
```

```python
embark=pd.get_dummies(t['Embarked'],drop_first=True)
embark.head(5)
```

```python
pclass=pd.get_dummies(t['Pclass'],drop_first=True)
pclass.head(5)
```

```python
t=pd.concat([t,sex,embark,pclass],axis=1)
```

```python
t.head()
```

```python
t.drop(['PassengerId','Pclass','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
t.head()
```

# Train Data

```python
x=t.drop('Survived',axis=1)
y=t['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)*100
```
